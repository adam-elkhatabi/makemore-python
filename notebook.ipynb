```python
x = np.array([1, 2, 3, 4])

y1 = np.array([10, 9, 7, 16])
y2 = np.array([1, 15, 20, 3])
y3 = np.array([10, 10, 10, 10])

plt.plot(x, y1)
plt.plot(x, y2)
plt.plot(x, y3)

plt.title("Class Size", fontsize=20, family='Arial', fontweight="bold", color="#d6351c")


plt.xlabel("grade", fontsize=10, family="Arial", fontweight="bold", color="#ba2fd6")
plt.ylabel("size", fontsize=10, family="Arial", fontweight="bold", color="#ba2fd6")

plt.xticks(x)

plt.tick_params(axis="both", colors="#ba2fd6")

plt.grid(axis='y', linestyle="dashed", linewidth=2)

plt.show()
```

## Bar charts

## Pie Charts

## Scatter Chart

```python
X = [[3.0, 1.0],
     [1.0, 2.5],
     [0.5, 1.0],
     [1.0, 1.0]]

y = [1.0, -1.0, -1.0, 1.0]

for xi, yi in zip(X, y):
    color = "red" if yi == 1.0 else "blue"
    plt.scatter(xi[0], xi[1], color=color, edgecolors='k', s=100)

plt.title("Toy Classification DataSet (Red = 1.0 and Blue = -1.0)")
plt.xlabel("Feature1")
plt.ylabel("Feature2")

plt.grid()

plt.show()
```

## Histogram

```python
g = np.random.default_rng(seed=101)

scores = g.normal(loc=80, scale=2, size=100)
scores = scores.clip(0, 100)

plt.hist(scores)

plt.show()
```
